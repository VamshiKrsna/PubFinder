```python
import pandas as pd
import numpy as np

df = pd.read_csv("open_pubs.csv")

df
```

```python
df.columns
```

```python
df.columns = ['idx','Name','Address','PostalCode','Easting','Northing','Latitude','Longitude','LocalAuthority']
```

```python
df
```

```python
df.info()
```

```python
df['Latitude'] = df['Latitude'].replace(r'\N',np.nan)
df.Latitude.value_counts()
```

```python
df['Longitude'] = df['Longitude'].replace(r'\N',np.nan)
df.Latitude.value_counts()
```

```python
df['Latitude'] = df['Latitude'].astype('float')
df['Longitude'] = df['Longitude'].astype('float')
```

```python
df['Latitude'] = df['Latitude'].dropna()
```

```python
df['Longitude'] = df['Longitude'].dropna()
```
